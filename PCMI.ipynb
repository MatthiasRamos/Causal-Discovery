```
# Imports
import numpy as np
import matplotlib
from matplotlib import pyplot as plt   
import sklearn

import tigramite
from tigramite import data_processing as pp
from tigramite import plotting as tp
from tigramite.pcmci import PCMCI
from tigramite.independence_tests import ParCorr, GPDC, CMIknn, CMIsymb
from tigramite.models import LinearMediation, Prediction
```

```
import pandas as pd
df = pd.read_csv('air-data/PRSA_Data_20130301-20170228/PRSA_Data_Aotizhongxin_20130301-20170228.csv')

df = df.fillna(0)
df = df.drop(['No', 'year', 'month', 'day', 'hour', 'station'], axis=1)
#remplacement des directions du vent en valeurs discrètes
df = df.replace({'N' : 0, 'NNW' : 1, 'NW' : 2, 'WNW' : 3, 'W': 4, 'WSW' : 5 ,'SW' : 6, 'SSW' : 7, 'S' : 8, 'SSE': 9, 'SE':10, 'ESE': 11, 'E': 12, 'ENE': 13, 'NE': 14, 'NNE':15})

df.to_csv("air-data/transformed_data.csv", index=False)
```

```
# Initialize dataframe object, specify time axis and variable names
data = pd.read_csv("air-data/transformed_data.csv")
data = data.drop('RAIN', axis=1)
data = np.array(data)[:1000,:]
dataframe = pp.DataFrame(data, )
```

```
import time
start = time.time()
cmi_knn = CMIknn(significance='shuffle_test', knn=0.2, shuffle_neighbors=5, transform='ranks')
pcmci_cmi_knn = PCMCI(
    dataframe=dataframe, 
    cond_ind_test=cmi_knn,
    verbosity=2)
results = pcmci_cmi_knn.run_pcmci(tau_max=3, pc_alpha=0.05)
print(f'exec time : {time.time()-start}')
```

```
pcmci_cmi_knn._print_significant_links(
        p_matrix = results['p_matrix'], 
        val_matrix = results['val_matrix'],
        alpha_level = 0.05)
```
